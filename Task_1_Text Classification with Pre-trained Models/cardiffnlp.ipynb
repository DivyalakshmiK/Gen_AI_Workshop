```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
model_name = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
def classify(sentence):
  inputs = tokenizer(sentence, return_tensors="pt")
  output = model(**inputs)

  scores = output.logits[0].detach().numpy()
  predicted = scores.argmax()
  labels = ["Negative", "Neutral","Positive"]
  return labels[predicted]
```

```python
sentences = ["This is a terrible place", "It okay, I guess", "I finally got into my dream university!", "The server is down and great I lost my progress", "it's okayish i guess, not better not worst"]

for i in sentences:
  print(f"{i} - {classify(i)}")
```
