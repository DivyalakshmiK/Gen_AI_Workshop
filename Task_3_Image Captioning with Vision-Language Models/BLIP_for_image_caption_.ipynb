```python
import torch
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")
```

```python
path = "bow down.jpg"
image = Image.open(path).convert("RGB")
```



> Created a processor / image encoder and passed on the image, the result is then fed to the generator for related text generation (Caption). Then the generator results are decoded using the processor, decoder.



```python
inputt = processor(images = image, return_tensors = "pt")
generated_ids = model.generate(**inputt, max_length = 20)

caption = processor.decode(generated_ids[0], skip_special_tokens = True)
print(caption)
```
